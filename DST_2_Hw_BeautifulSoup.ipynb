```python
data = []
url = 'https://www.hse.ru/org/persons/?ltr=%D0%90;udept=22726'
r = requests.get(url)

if r.status_code == 200:
    soup = BeautifulSoup(r.text, 'lxml')
    names = soup.findAll('div', class_ = 'post__content post__content_person')
    
    for i in names:
        people_name = i.find('div', class_="content__inner content__inner_foot1").find('a').text.strip()
        pos = i.find('p',class_="with-indent7").find('span').text.strip().split(':' )
        position = pos[0]
        
        try: 
          w = pos[1].strip().split('/')
          w0 = w[0].strip()
          if len(w)>1:
            w1 = w[1].strip()
            work_place = w0 +' / '+ w1
          else:
            work_place = w0
        except: 
          work_place = ' '
        
        try: 
          phone = i.find('div', class_ = 'l-extra small').find('span').text
        except: 
           phone = ' '
        
        try: 
          e_mail = str(i.find('div', class_='l-extra small').find('a')).split('[')[1].split(']')[0].replace('"', '').replace(',', '').replace('-at-', '@')
        except: 
           e_mail = ' '
        personal_account = 'https://www.hse.ru' + i.find('a', class_ = 'link link_dark large b').get('href')
  
        data.append([people_name, position, work_place, phone, e_mail, personal_account])
        
else:
    print("Ошибка")
       
header = ['people_name', 'position', 'work_place', 'phone', 'e_mail', 'personal_account']
df = pd.DataFrame(data, columns = header)
        
```

```python
# Создаем функцию, принимающую на вход ссылку на аккаунт преподавателя, возвращающую строчку с началом года работы.
def person(url): 
  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'lxml')
  try: 
    year = soup.find('li', class_ = 'i').text
  except: 
    year = ' '
        
  return year
```

```python
url = 'https://www.hse.ru/org/persons/224839542'
person(url)
```

```python
# Создаем новую колонку в df.
df.personal_account
df['work`s_year'] = df.personal_account.apply(person)
df
```

```python
# Применение цикла For.
list_ = []
for i, url in enumerate(df.personal_account):
  list_.append(person(url))
  print(i)

df['work`s_year(2)'] = list_

  
```
